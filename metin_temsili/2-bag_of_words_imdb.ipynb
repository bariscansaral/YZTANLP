```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
df=pd.read_csv("IMDB Dataset.csv")
display(df)
```

```python
reviews=df["review"]
print(reviews)
sentiment=df["sentiment"]
print("\n\n",sentiment)
```

```python
for i in range(0,4):
    print("\n\n",reviews[i])
```

```python
cleaned_reviews=reviews.copy()
for i in range(len(cleaned_reviews)):
    cleaned_reviews[i] = cleaned_reviews[i].lower()
    cleaned_reviews[i] = re.sub(r"\d+","",cleaned_reviews[i])
    cleaned_reviews[i] = re.sub(r"[^\w\s]","",cleaned_reviews[i])
    cleaned_reviews[i] = re.sub(r"<.*?>", "", cleaned_reviews[i])
    cleaned_reviews[i] = " ".join([word for word in cleaned_reviews[i].split() if len(word)>2])
    cleaned_reviews[i] = " ".join([word for word in cleaned_reviews[i].split() if word not in stop_words])
    
```

```python
print(reviews.head())
print("\n\n",cleaned_reviews.head())
```

```python
bow_model=CountVectorizer()
review_vectors=bow_model.fit_transform(cleaned_reviews[:100])
print(review_vectors)
```

CountVectorizer çıktısı sparse (seyrek) bir matristir ve her (satır, sütun) çifti bir kelimenin ilgili dokümanda kaç kez geçtiğini gösterir. “Coords” kısmı (doküman indexi, kelime indexi) konumunu, “Values” ise bu kelimenin o doküman içinde kaç kez geçtiğini ifade eder.

Örnek: (0, 485) = 6 → 0. review içinde vocabulary’de 485 indexine karşılık gelen kelime **o tek review içinde 6 kez** geçmiştir. Yani tüm dataset içinde değil, sadece 0. satırdaki (0. review) metninde sayılır.

Benzer şekilde (99, 645) = 1 → 99. review içinde 645 indexli kelime **o review’e özel olarak 1 kez** geçmiştir.

Benzer şekilde (33,485) = 8 -> 33. review içinde 485 indexli kelime (iki üstteki kelime mesela use olsun) **33. reviewde 8 kere geçmişken 0. reviewde 6 kez geçmiştir** gibi

```python
#Indexlenmiş kelimeler:
vocab=bow_model.get_feature_names_out()
print(vocab)
```

```python
#Kelimeler ve index karşılıkları
index_to_word = {i: word for i, word in enumerate(vocab)}
print(index_to_word)
```

```python
word_counts=review_vectors.sum(axis=0).A1
print(word_counts)
word_freq = {index_to_word[i]: int(word_counts[i]) for i in range(len(word_counts))}
print(word_freq)
```

```python
most_5=Counter(word_freq).most_common(5)
print("\nEn çok geçen 5 kelime",most_5)
```
