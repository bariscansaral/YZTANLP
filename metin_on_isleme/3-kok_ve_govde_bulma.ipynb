```python
import nltk
nltk.download("wordnet")
nltk.download("omw-1.4")
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
#stem, kök bulma
words_stem=["playing","played","plays","happier","happily","studying","studies","learning","learnt","coming","come","came","smiling","smiled"]
stems=[stemmer.stem(i) for i in words_stem]
print(f"Orijinal kelimeler: {words_stem}")
print(f"Kökler: {stems}")
```

```python
#Lemma, gövde bulma
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
words_lemma=["running","ran","gone","better","children","cards","peoples","studies","coming"]
lemmas=[lemmatizer.lemmatize(i) for i in words_lemma]
print("Orijinal Kelimeler: ",words_lemma)
print("Gövdeler: ",lemmas)
```
