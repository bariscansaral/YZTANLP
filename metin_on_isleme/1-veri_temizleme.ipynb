```python
import string
import re
```

```python
raw_text="Google     NLP        Dersi"
print(raw_text)
normalized_text1=" ".join(raw_text.split())
print(normalized_text1)
```

```python
raw_text2="GooGle NlP DerSi"
print(raw_text2)
normalized_text2=raw_text2.lower()
print(normalized_text2)
```

```python
raw_text3="AI Natural-Language-Processing"
print(raw_text3)
normalized_text3=raw_text3.translate(str.maketrans(" "," ",string.punctuation))
print(normalized_text3)
```

```python
raw_text4="Natural @ Language > Processing % #"
print(raw_text4)
normalized_text4=re.sub(r"[^A-Za-z0-9\s]","e",raw_text4)
print(normalized_text4)
```

```python
from textblob import TextBlob
raw_text5="İt is amaszing in 2066"
print(raw_text5)
normalized_text5=TextBlob(raw_text5).correct()
print(normalized_text5)
```

```python
from bs4 import BeautifulSoup
raw_html="<div> 2025 Google <div>"
print(raw_html)
normalized_html=BeautifulSoup(raw_html,"html.parser").get_text()
print(normalized_html)
```
